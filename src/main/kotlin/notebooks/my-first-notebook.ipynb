Connect to the database and load the entire contents into memory. This does defeat the purpose of creating a database to begin with, so I would not recommend making this your standard workflow (especially if you have more than 100,000 rows), but for our exploratory purposes it's ok for now.

You can load the data directly from the notebook by passing in the database config parameters (url, username and password)  and using the `readSqlTable` function. I have defined the config parameters in the `util` package, so we can use them in multiple notebooks, but of course you could define them here. If the `util` constants didn't import to the notebook, make sure you go to the settings for this notebook and `select modules to use in the notebook`

```kotlin
import util.URL
import util.USER_NAME
import util.PASSWORD

val dbConfig = DbConnectionConfig(URL, USER_NAME, PASSWORD)
val data = DataFrame.readSqlTable(dbConfig, "powerlifting_data")
```

Now that we have loaded the data, we can inspect it by using the `describe` method from kotlin dataframe
